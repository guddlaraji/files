```python
import numpy as np
a = np.array([1,2,3,4,5,6])
print(a)
```

```python
#convert the above array into series
sr = pd.Series(a)
print(sr)
print(sr.shape)
```

**2D-DataFrame**

```python
#Dataframe is a 2D size ,Heterogenous data with labelled axis(rows and columns)
#columns also known as --> features,varaibles, fields
#rows also known as --> records, values and index
```

```python
#what is the application of DataFrame?
#works on datasets
#analysis
#dropping
#processing
#cleaning
#join the multiple data
#create the excel, json..etc
#compute the mathematical and statistical operations
#use group by function
```

```python
#create a df from the list of numbers
l = [1,2,3,4,5,6]
df = pd.DataFrame(l)
print(df)
df.shape
```

```python
#create a dataframe from the set
s1 = {'apple','banana','orange','pineapple'}
df = pd.DataFrame(s1,index=['a','b','c','d'])
print(df)
#Note:- we can create a df from the SET
```

```python
#create a dataframe from the dictionary
d = {
    "name":'medha',
    "course":"DS",
    "Duration":"4months"
}

df = pd.DataFrame(d,index=['a','b'])
print(df)
```
