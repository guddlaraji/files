```python
import pandas as pd


candidates = {
    'gmat': [780,750,690,710,680,730,690,720,740,690],
    'gpa': [4.0,3.9,3.3,3.7,3.9,3.7,2.3,3.3,3.3,1.7],
    'work_experience': [3,4,3,5,4,6,1,4,5,1],
    'admitted': [1,1,0,1,0,1,0,1,1,0]
}
```

```python
df = pd.DataFrame(candidates)
print(df.head())
```

**Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,5))
sns.scatterplot(x='gmat', y='gpa', hue='admitted', data=df, s=120)
plt.title("GMAT vs GPA (Colored by Admission Result)")
plt.show()
```

**Split Data (Train/Test)**

```python
from sklearn.model_selection import train_test_split

# Features and Target
X = df[['gmat', 'gpa', 'work_experience']]
y = df['admitted']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

**Build & Train Logistic Regression Model**

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)

print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)
```

```python
new_data = [[650, 3.2, 2]]
prediction = model.predict(new_data)

print("\nPrediction for new candidate:", prediction[0])
```
